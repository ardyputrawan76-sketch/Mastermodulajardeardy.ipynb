```python
import os
import time
from google import genai
from google.genai import types
from google.genai import errors
from docx import Document
from google.colab import userdata

# 1. AMBIL API KEY DARI SECRETS COLAB
api_key_colab = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key_colab)

# 2. TEMPLATE PROMPT MATANG
PROMPT_TEMPLATE = """
Kamu adalah seorang ahli kurikulum pendidikan Indonesia dan guru penggerak yang sangat kompeten.
Tugasmu adalah menyusun sebuah Modul Ajar Kurikulum Merdeka yang lengkap, sistematis, dan siap pakai untuk keperluan pemeriksaan administrasi sekolah (supervisi).

Modul ini HARUS mengintegrasikan Pendekatan **Deep Learning** (pembelajaran mendalam) yang terdiri dari unsur **3M: Mindful, Meaningful, dan Joyful Learning**.

Berikut adalah variabel data yang harus kamu gunakan:
- Mata Pelajaran: Pendidikan Jasmani, Olahraga, dan Kesehatan (PJOK)
- Kelas: {kelas}
- Materi Pokok: {nama_materi}

Hasilkan output Modul Ajar dengan mengikuti KERANGKA RESMI berikut ini secara berurutan. Jangan melewatkan satu sub-bab pun.

---

### I. INFORMASI UMUM
- **Identitas Modul:** Nama Penyusun (Bapak Guru), Institusi (SD Negeri), Tahun Pelajaran (2026/2027).
- **Kompetensi Awal:** Tuliskan 2 kemampuan mendasar yang harus dimiliki siswa sebelum mempelajari materi {nama_materi} ini.
- **Profil Pelajar Pancasila:** Pilih 2-3 dimensi yang paling relevan (misal: Mandiri, Gotong Royong, Bernalar Kritis) dan jelaskan singkat bagaimana penerapannya dalam materi {nama_materi}.
- **Sarana dan Prasarana:** Sebutkan alat, media pembelajaran, atau lapangan yang rasional digunakan untuk materi ini di jenjang kelas {kelas}.
- **Target Peserta Didik:** Peserta didik reguler/tipikal.
- **Model Pembelajaran:** Sebutkan model yang cocok (misal: Tatap Muka / Problem Based Learning / Project Based Learning).

### II. KOMPONEN INTI
- **Tujuan Pembelajaran (TP):** Rumuskan TP yang jelas, terukur, dan mencakup ranah kognitif, afektif, dan psikomotorik yang sesuai dengan tingkat kematangan kelas {kelas}.
- **Pemahaman Bermakna (Mindful Basis):** Manfaat nyata jangka panjang bagi kehidupan dan kesehatan siswa setelah menguasai materi {nama_materi} ini.
- **Pertanyaan Pemantik:** Buat 2 pertanyaan terbuka yang menantang nalar siswa di awal kelas terkait materi ini.
- **Kegiatan Pembelajaran:**
  * **A. Kegiatan Pendahuluan:** (Apersepsi, motivasi, penyampaian tujuan, dan pemanasan yang relevan).
  * **B. Kegiatan Inti (Wajib Menerapkan Pendekatan Deep Learning - 3M secara runtut):**
    1. **Tahap Mindful (Kesadaran):** Rancang aktivitas di mana guru membangun kesadaran emosional dan logika siswa mengenai *apa* dan *mengapa* gerakan/materi {nama_materi} ini penting bagi tubuh mereka. Siswa diajak mengamati masalah atau contoh awal.
    2. **Tahap Meaningful (Kebermaknaan):** Rancang aktivitas eksplorasi tingkat tinggi (HOTS). Siswa dibagi kelompok untuk berdiskusi, menganalisis teknik, saling mengoreksi gerakan (peer-assessment), atau memecahkan tantangan gerak secara mandiri. Fokus pada kedalaman pemahaman teknik, bukan hafalan teori.
    3. **Tahap Joyful (Menyenangkan):** Rancang aktivitas praktik langsung yang dikemas dalam bentuk permainan modifikasi (games-based learning), kompetensi kelompok yang seru, atau tantangan berhadiah (ice breaking gerak) agar emosi positif siswa terbangun saat melakukan aktivitas fisik materi {nama_materi}.
  * **C. Kegiatan Penutup:** (Pendinginan, refleksi bersama, kesimpulan bersama siswa, evaluasi singkat, dan doa).
- **Asesmen (Penilaian):** Buat secara deskriptif rencana penilaiannya (Sikap dan Performa).
- **Pengayaan dan Remedial:** Rencana instruksional singkat untuk siswa yang cepat paham dan siswa yang butuh bimbingan ekstra.
- **Refleksi Peserta Didik dan Guru:** Berikan 3 pertanyaan evaluasi untuk mengukur keberhasilan pembelajaran dari sudut pandang siswa dan guru.

### III. LAMPIRAN
- **Lembar Kerja Peserta Didik (LKPD):** Rancang instruksi tugas kelompok/individu yang menantang dan mencerminkan tahap Meaningful tadi.
- **Bahan Bacaan Guru & Peserta Didik:** Ringkasan materi esensial mengenai teknik dasar {nama_materi} kelas {kelas} dalam 2-3 paragraf.
- **Bank Soal Evaluasi:** Buatkan 10 soal pilihan ganda (lengkap dengan pilihan A, B, C, D) dan 5 soal essay. Semua soal harus berbasis HOTS (Higher Order Thinking Skills) sesuai materi {nama_materi} kelas {kelas}.
- **Kunci Jawaban & Rubrik:** Berikan kunci jawaban yang jelas untuk pilihan ganda, serta pedoman penskoran detail untuk soal essay.
- **Glosarium:** Daftar kata/istilah teknis penting dalam materi ini beserta artinya.
- **Daftar Pustaka:** Sumber referensi (Buku Guru, Buku Siswa Kurikulum Merdeka Kemendikbudristek).

---
Catatan untuk AI: Hasilkan teks langsung dalam bahasa Indonesia yang formal, terstruktur rapi, tanpa ada teks pembuka atau penutup basa-basi dari AI (langsung ke isi dokumen).
"""

def generate_modul_ajar(kelas, nama_materi):
    print(f"Sedang memproses Modul Ajar untuk Kelas {kelas} - Materi: {nama_materi}...")

    config_premium = {
        "max_output_tokens": 8000,
        "temperature": 0.7
    }

    prompt_siap = PROMPT_TEMPLATE.format(kelas=kelas, nama_materi=nama_materi)

    # KODE PINTAR: Mengulang otomatis jika server gratisan sedang penuh
    for percobaan in range(5):
        try:
            response = client.models.generate_content(
                model='gemini-2.5-flash',
                contents=prompt_siap,
                config=config_premium
            )
            return response.text
        except errors.APIError as e:
            print(f"Jalur gratis sibuk, otomatis mengantre... (Mencoba ulang ke-{percobaan+1}/5)")
            time.sleep(5)
            continue
        except Exception as e:
            raise e

    raise RuntimeError("Server Google sangat padat saat ini, silakan klik Play lagi semenit kemudian.")

def save_to_word(text_content, filename):
    doc = Document()
    for line in text_content.split('\n'):
        if line.startswith('### '):
            doc.add_heading(line.replace('### ', ''), level=1)
        elif line.startswith('- '):
            doc.add_paragraph(line, style='List Bullet')
        elif line.strip() == '---':
            doc.add_page_break()
        else:
            doc.add_paragraph(line)

    doc.save(filename)
    print(f"Selesai! File berhasil dibuat: {filename}")

# --- INPUT DATA DISINI ---
input_kelas = "6"
input_materi = "Kombinasi Gerak Dasar Passing Bawah Bola Voli"
# -------------------------

nama_file = f"Modul_Ajar_Kelas_{input_kelas}_{input_materi.replace(' ', '_')}.docx"

# EKSEKUSI PROGRAM
hasil_modul = generate_modul_ajar(input_kelas, input_materi)
save_to_word(hasil_modul, nama_file)
```
